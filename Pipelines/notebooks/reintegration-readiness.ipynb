# Reintegration Readiness

## Problem Framing
Predict resident reintegration readiness to support case planning and intervention prioritization.

## Data Acquisition, Preparation & Exploration
Pipeline implementation: `ML_Pipeline_3/run_pipeline.py`, `ML_Pipeline_3/src/features.py`, `ML_Pipeline_3/src/modeling.py`.

## Modeling & Feature Selection
Readiness models generate continuous scores and tier labels for resident monitoring.

## Evaluation & Interpretation
Evaluation results are interpreted in terms of case conference and reintegration decision support.

## Causal and Relationship Analysis
Explanations identify influential factors while clearly noting observational-data limitations.

## Deployment Notes
Integrated through `backend/Services/MlScoringService.cs` and exposed in `backend/Controllers/MlInsightsController.cs`.
