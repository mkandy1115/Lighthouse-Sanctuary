# Social Engagement Referrals

## Problem Framing
Predict which social content is most likely to drive meaningful engagement and donation referrals.

## Data Acquisition, Preparation & Exploration
Pipeline implementation: `ML_Pipeline_2/run_pipeline.py`, `ML_Pipeline_2/src/features.py`, `ML_Pipeline_2/src/modeling.py`.

## Modeling & Feature Selection
Engagement and referral models are trained and combined into uplift/churn-oriented post scores.

## Evaluation & Interpretation
Model quality is evaluated for operational use in social strategy decisions.

## Causal and Relationship Analysis
Observed relationships are interpreted carefully as predictive signals, not guaranteed causality.

## Deployment Notes
Integrated through `backend/Services/MlScoringService.cs` and reported via `backend/Controllers/MlInsightsController.cs`.
