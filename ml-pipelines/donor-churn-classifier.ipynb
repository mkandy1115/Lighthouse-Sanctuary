# Donor Churn Classifier

## Problem Framing
Predict donor churn risk so fundraising staff can prioritize retention outreach. Goal type is predictive with explanatory interpretation support.

## Data Acquisition, Preparation & Exploration
Pipeline implementation: `ML_Pipeline_1/run_pipeline.py`, `ML_Pipeline_1/src/features.py`, `ML_Pipeline_1/src/modeling.py`.

## Modeling & Feature Selection
Model training and selection occur in `train_and_evaluate(...)`, then supporters are scored.

## Evaluation & Interpretation
Evaluation metrics are used to choose production scoring behavior and classify risk tiers.

## Causal and Relationship Analysis
Feature influence is discussed for operations, while avoiding unsupported causal claims.
